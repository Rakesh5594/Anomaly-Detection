```python
import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
```

```python
df = pd.read_csv('Sample scores.csv')
df
```

```python
model=IsolationForest(n_estimators=1000,max_samples='auto',contamination=float(0.2),max_features=1.0)
model.fit(df[['Scores']])
```

```python
df['scores']=model.decision_function(df[['Scores']])
df['anomaly']=model.predict(df[['Scores']])
df.head(20)
```

```python
outliers_counter = len(df[df['Scores'] > 36])
outliers_counter
```
